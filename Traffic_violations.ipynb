```python
#initialize variables for empty lists to hold data

date_stop_ls = [] 
time_stop_ls = [] 
agency_ls = [] 
subagency_ls = [] 
description_ls= [] 
location_ls = [] 
latitude_ls = [] 
longitude_ls = []
vehicle_type_ls = []
vehicle_year_ls = []
vehicle_make_ls = []
vehicle_model_ls = []
vehicle_color_ls = []
vehicle_type_ls = []
violation_type_ls = []
race_ls = []
gender_ls = []
driver_state_ls = []
driver_lic_st_ls = []
arrest_type_ls = []
```

```python
#fill lists with data

for violations in data['data']:  #data[index]
    
    #add info to list for each record
    date_stop_ls.append(violations[9])
    time_stop_ls.append(violations[10])
    agency_ls.append(violations[11])
    subagency_ls.append(violations[12])
    description_ls.append(violations[13])
    location_ls.append(violations[14])
    latitude_ls.append(violations[15])
    longitude_ls.append(violations[27])
    vehicle_type_ls.append(violations[28])
    vehicle_year_ls.append(violations[29])
    vehicle_make_ls.append(violations[30])
    vehicle_model_ls.append(violations[31])
    vehicle_color_ls.append(violations[32])
    violation_type_ls.append(violations[33])
    race_ls.append(violations[37])
    gender_ls.append(violations[38])
    driver_state_ls.append(violations[40])
    driver_lic_st_ls.append(violations[41])
    arrest_type_ls.append(violations[42])
```

```python
import pandas as pd
```

```python
violation_ls = list(zip(date_stop_ls, time_stop_ls, agency_ls, subagency_ls, description_ls, location_ls,
                        latitude_ls, longitude_ls, vehicle_type_ls, vehicle_year_ls, vehicle_make_ls, 
                        vehicle_model_ls, vehicle_color_ls, violation_type_ls, race_ls, gender_ls, 
                        driver_state_ls, driver_lic_st_ls, arrest_type_ls))
```

```python
viol_col = ['StopDate','StopTime', 'Agency', 'Subagency', 'Description', 'Location', 'Latitude', 'Longitude',
            'VehicleType', 'VehicleYear', 'VehicleMake', 'VehicleModel', 'VehicleColor', 'ViolationType',
            'Race', 'Gender', 'DriverState', 'DriverLicSt', 'ArrestType']
```

```python
viol_df = pd.DataFrame(violation_ls, columns = viol_col)
```
